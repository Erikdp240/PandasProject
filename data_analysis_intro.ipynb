# Advanced Data Selection

```python
data[data['year'] > 1995] # shows movies made after 1995
```

```python
data[data['year'] == 1966] # shows movies only made in 1966
```

```python
data[(data['year'] > 1995) & (data['year'] < 2000)] # shows data between those years using AND
```

```python
data[(data['year'] > 1995) | (data['year'] < 2000)] # shows data using OR
```

```python
data[(data['year'] > 1995) & (data['year'] < 2000)].sort_values(by='rating', ascending=False).head(10)
# shows data between those years using AND
# sorts the values by descending order by rating
# shows the first 10 results
```

# Grouping

```python
data.groupby(data['year'])['rating'].mean() # tells you the average rating for movies by year
```

```python
data.groupby(data['year'])['rating'].max() # tells you the max rating for movies by year
```

```python
data.groupby(data['year'])['rating'].min() # tells you the min rating for movies by year
```
